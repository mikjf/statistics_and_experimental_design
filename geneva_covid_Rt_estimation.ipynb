# Covid-19 $R_t$ in real-time (Geneva region)

In this exercise adapted from [Kevin Systrom repository](https://github.com/k-sys/covid-19/blob/master/Realtime%20R0.ipynb) and [Constructor Learning](https://learning.constructor.org/), we'll measure $R_t$, the number of people who become infected per infectious person at time t. The basic reproduction number is $R_0$ at t=0, the average number of secondary infections produced by a typical case of an infection in a population when everyone is susceptible.

The basic reproductive number is affected by several factors:

* The rate of contacts in the host population
* The probability of infection being transmitted during contact
* The duration of infectiousness

**Note**: the code below is taken from the author repository mentioned above. The goal is to understand the procedure and apply it for the data relevant for our Geneva region.

## 1. Poisson distribution function approach

### 1.1 Choosing a likelihood function $P\left(k_t|R_t\right)$

How is it likely to have k new cases give a value or Rt?

```python
# 'arrivals' model -> poisson distribution
# Î» = average new cases per day
# probability of seeing k new cases is distributed with a poisson distribution
```

$$P(k|\lambda) = \frac{\lambda^k e^{-\lambda}}{k!}$$

```python
# import packages
import numpy as np
import pandas as pd
from scipy.stats import poisson
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create vector for k (0 to 69 with step 1)
k = range(70)

# create list of different lambdas
lambdas = [10, 20, 30, 40]

# initialize empty matrix
y = np.zeros((len(k), len(lambdas)))
```

```python
# probability mass function (PMF) for every lambda (poisson is discrete)
for i in range(len(lambdas)):
    y[:,i] = poisson.pmf(k, lambdas[i])
    
print(y.shape)
```

```python
# create pmf dataframe
pmf_df = pd.DataFrame(y, columns=['k10', 'k20', 'k30', 'k40'])
pmf_df.head()
```

```python
# set sns theme
sns.set_theme()
```

```python
# plot

# plot for each lambda
ax = pmf_df.plot(marker = 'o',
                 linestyle = 'None',
                 fontsize = 14,
                 figsize = (13,9),
                 color = ['red', 'orange', 'yellow', 'blue'],
                 xticks = [0, 20, 40, 60]
                 )

# set y label
ax.set_ylabel("p(k|lambda)", fontdict = {'fontsize': 20}, labelpad = 20)

# set legend and legend title
leg = ax.legend(fontsize = 16, title= 'lambda', labels = [lab[1:] for lab in pmf_df.columns])
leg.set_title('lambda', prop = {'size':18})

# set plot title
ax.set_title("Poisson distribution of Cases p(k|lambda)", pad = 15, fontdict = {'size':22});
```

### 1.2 Connecting lambda and Rt

We have $P\left(\lambda_t|k_t\right)$ which is parameterized by $\lambda$
but we were looking for $P\left(k_t|R_t\right)$ which is parameterized by $R_t$.
We need to know the relationship between $\lambda$ and and $R_t$.

Since we know every new case count on the previous day, we can now reformulate
the likelihood function as a Poisson parameterized by fixing $k$ and varying $R_t$.

$$P\left(k|R_t\right) = \frac{\lambda^k e^{-\lambda}}{k!}$$

### 1.3 Evaluating the likelihood function

What is the likelihood of different value of Rt on each of the days that have k new case counts?

```python
# given a set of new case counts k list
k = np.array([20, 40, 55, 90])
```

```python
# create array for all possible values of Rt
R_T_MAX = 12
r_t_range = np.linspace(start = 0, stop = R_T_MAX, num = R_T_MAX*100+1)
```

```python
# set gamma (1/serial interval)
GAMMA = 1/7
```

```python
# map rt into lambda
# !!! N-1 lambdas because on the first day of an outbreak you don't know what to expect
lam = np.dot(k[:-1].reshape(-1,1), np.exp(GAMMA * (r_t_range - 1).reshape(1,-1)), )

# evaluate the likelihood on each day and normalize sum of each day to 1.0
likelihood_r_t = np.zeros((lam.shape[1], len(k)-1)) # initialize an empty matrix  

for i in range(len(k)-1):
    likelihood_r_t[:,i] = poisson.pmf(k[1:][i], lam[i])

# sum of the elements of vector (along rows)
lh = np.sum(likelihood_r_t, axis = 0)

# normalization 
likelihood_r_t = (lambda x: x/lh)(likelihood_r_t)

# plot dataframe
likelihood_df = pd.DataFrame(likelihood_r_t, r_t_range).reset_index()
likelihood_df.columns = ['Rt', 'k40', 'k55', 'k90']

# plot by different k
ax = likelihood_df.set_index('Rt').plot(
    fontsize = 14,
    figsize = (10,8),
    color = ['red', 'orange', 'yellow'],
    lw = 7,
    xlim = (0, 10),
    xticks = [0, 2.5, 5, 7.5, 10]
    )

# set labels, legend, titles
ax.set_ylabel("L(Rt | k)", fontdict = {'fontsize': 20}, labelpad = 20)
ax.set_xlabel("Rt", fontdict = {'fontsize': 20}, labelpad = 20)
leg = ax.legend(fontsize = 16, labels = [40, 55, 90])
leg.set_title('New cases', prop = {'size':18})
ax.set_title("Likelihood of Rt given k", pad = 15, fontdict = {'size':22})
ax;
```

- for each day we have an independent guess for Rt
- we know that Rt today depends on Rt-1 and so on
- bayesian statistics let us incorporate prior knowledge into parameter estimation

## 2. Bettencourt & Ribeiro's approach

Having seen k new cases, we believe that the distirbution Rt is equal to, **for a single day**:

- the likelihood of seeing k new cases given Rt
- multiplied by
- the prior belief of the value of P(Rt) without the data
- divided by
- the probability of seeing this many cases in general

For multiple days, we'll assume that the distribution of Rt is Gaussian centered around Rt-1.

### 2.1 Bayesian update (combine the information in the previous days with current day)

- bayesian update -> multiply the likelihood by the prior (previous day) to get the posteriors
- cumulative product of each successive day

```python
# apply cumulative product to likelihood (along the columns)
posteriors = np.cumprod(likelihood_r_t, axis = 1) 

# sum of the elements of vector (along rows)
ps = np.sum(posteriors, axis = 0)

# normalization 
posteriors = np.apply_along_axis(lambda x:x/ps, 1, posteriors) 
```

```python
# plot

# posteriors dataframe
posteriors_df = pd.DataFrame(posteriors, r_t_range).reset_index()
posteriors_df.columns = ['Rt', 'd1', 'd2', 'd3']

# plot by different days
ax = posteriors_df.set_index('Rt').plot(
    fontsize = 14,
    figsize = (10,8),
    color = ['red', 'orange', 'yellow'],
    lw = 7,
    xlim = (0, 10),
    xticks = [0, 2.5, 5, 7.5, 10]
    )

# set labels, legend, titles
ax.set_ylabel("P(Rt | k)", fontdict = {'fontsize': 20}, labelpad = 20)
ax.set_xlabel("Rt", fontdict = {'fontsize': 20}, labelpad = 20)
leg = ax.legend(fontsize = 16, labels = [1, 2, 3])
leg.set_title('Day', prop = {'size':18})
ax.set_title("Posterior of Rt given k", pad = 15, fontdict = {'size':22})
ax;
```

- notice how day 1 matches the first likelihood in the previous plot as no information other than that day is used
- the more the days the narrower the distribution

### 2.2 What's the most likely value or Rt each day?

```python
# most likely values
mlv_day1, mlv_day2, mlv_day3 = [round(posteriors_df.set_index('Rt')[day].idxmax(), 3) for day in ['d1', 'd2', 'd3']]

# print values
print(
    "Most likely Rt values", "_"*20, "",
    f"Day 1: {mlv_day1}",
    f"Day 2: {mlv_day2}",
    f"Day 3: {mlv_day3}",
    sep = '\n'
)
```

## 4. Rt for GE region

The data for Switzerland has been retrieved from [here](https://raw.githubusercontent.com/openZH/covid_19/master/COVID19_Fallzahlen_CH_total.csv).

```python
# for 'positive' cases use column 'ncumul_conf'
```

```python
# get data from url
url = 'https://raw.githubusercontent.com/openZH/covid_19/master/COVID19_Fallzahlen_CH_total.csv'
```

```python
# get date, abbreviation_canton_and_fl, ncumul_conf
swiss = pd.read_csv(url, usecols = ['date', 'abbreviation_canton_and_fl', 'ncumul_conf'])
swiss.head()
```

```python
# convert date dtype object into datetime dtype
swiss['date'] = pd.to_datetime(swiss['date'], format = "%Y-%m-%d")

# sort by date
swiss.sort_values(by = 'date', inplace = True)

# rename columns
swiss.columns = ['date', 'canton', 'positive']

# visualize df
swiss
```

```python
swiss['canton'].unique()
```

```python
# prepare cases -> smoothing data
def prepare_cases(cases, cutoff):
    new_cases = np.diff(cases['positive'])
    smoothed = pd.Series(new_cases).rolling(7, center = True).mean()
    
    idx_start = smoothed.where(smoothed > cutoff).idxmin() #first index after cutoff
    smoothed_cutoff = smoothed.iloc[idx_start:len(smoothed)].fillna(method="ffill").values #from idx_start (incl) to end
    
    original = new_cases[idx_start:len(smoothed)]
    date = cases['date'].iloc[idx_start:len(smoothed)].values
    df_prepared = pd.DataFrame(
        {
            'smoothed': smoothed_cutoff,
            'original': original,
        },
        index = date
    )
    
    return df_prepared
```

```python
# filter swiss df
canton_name = 'GE'
cutoff = 5
cases = swiss.query('canton == "GE"')
```

```python
# get cases for TI
cases.head()
```

```python
df_prepared = prepare_cases(cases, cutoff)
df_prepared.head()
```

```python
# plot TI new cases per day
df_prepared.plot(
    fontsize = 14,
    figsize = (10,9),
    style = ['r-','k:'],
    legend = 'reverse',
    ).set_title(f"{canton_name} New Cases per Day", pad = 15, fontdict = {'size':22});
```

```python
# import norm
from scipy.stats import norm
```

```python
sigma=0.15
df_prepared = df_prepared.dropna()
sr = df_prepared['smoothed'].values # smoothed data obtained by rolling average

# (1) Calculate Lambda
lam = np.dot(sr[:-1].reshape(-1,1), np.exp(GAMMA * (r_t_range - 1).reshape(1,-1)), ) # taking k except last value

# (2) Calculate each day's likelihood

# (2a) Initialize an empty matrix  
likelihood_r_t = np.zeros((lam.shape[1], len(sr)-1)) # initialize an empty matrix  

# (2b) fill each column (each day) with likelihood
for i in range(len(sr)-1):
    likelihood_r_t[:, i] = poisson.pmf(round(sr[1:][i]), lam[i, :])

likelihood_df = pd.DataFrame(likelihood_r_t, r_t_range)#.reset_index()
likelihood_df.columns = df_prepared.index[1:]#.reset_index()['index'].tail(-1)
likelihood_df = likelihood_df.set_index(r_t_range)

# (3) Create the Gaussian Matrix
process_matrix = np.zeros((len(r_t_range), len(r_t_range)))

for i in range(len(r_t_range)):
    process_matrix[:,i] = norm.pdf(r_t_range, loc=r_t_range[i], scale=sigma)

# (3a) Normalize all rows to sum to 1
pm = np.sum(process_matrix, axis = 1) # sum of the elements of vector (along rows)
process_matrix = np.divide(process_matrix, pm) # normalization

# (4) Calculate the initial prior
prior0 = np.divide(([1]*len(r_t_range)), len(r_t_range))
#prior0 <- prior0/sum(prior0)

# Create a DataFrame that will hold our posteriors for each day
# Insert our prior as the first posterior.

posteriors = pd.DataFrame(np.zeros((len(r_t_range), len(df_prepared.index))))
posteriors.columns = df_prepared.index

posteriors.iloc[:,0] = prior0
```

```python
# (5) Iteratively apply Bayes' rule

for i in range(len(posteriors.columns)-1):
    previous_day = posteriors.columns[i]
    current_day  = posteriors.columns[i + 1]

    #(5a) Calculate the new prior
    current_prior = np.dot(process_matrix, posteriors[previous_day])
  
    #(5b) Calculate the numerator of Bayes' Rule: P(k|R_t)P(R_t)
    numerator = np.nan_to_num(likelihood_df[current_day]) * np.nan_to_num(current_prior)
  
    #(5c) Calcluate the denominator of Bayes' Rule P(k)
    denominator = sum(numerator)
    
    # Execute full Bayes' Rule
    posteriors[current_day] = np.divide(numerator, denominator)
```

```python
max_lh_idx = [posteriors[col].idxmax() for col in posteriors]
# collect all Rt values with corresponding indeces
Rt_calculated = [r_t_range[i] for i in max_lh_idx]
```

```python
Rt_df = pd.DataFrame(Rt_calculated).set_index(df_prepared.index)

p = Rt_df.plot(
    fontsize = 14,
    figsize = (14,9),
    style = ['ro-'],
    legend = None,
    markerfacecolor = 'k'
    ).set_title(f"{canton_name} Rt per Day", pad = 15, fontdict = {'size':22});

p.axes.autoscale(True)
p.axes.margins(0.035)
plt.axhline(y=1, color='blue', linestyle='-');
```
