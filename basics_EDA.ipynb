# Exploratory data analysis introduction

Here is an overview of how to perform simple EDA.

## 1. Load packages

```python
# import pandas package
import pandas as pd

# render figures in notebook
%matplotlib inline
```

## 2. Load data

```python
# ../ to go back
df = pd.read_csv("data/pw_stats_01_star_wars.csv")
df.head()
```

```python
# get information about dataframe
df.info()
```

## 3. Data wrangling

### 3.1 Select row based on condition

```python
# filter rows that are matching a specific value in a colum
df[df['species'] == 'Droid']
```

```python
# opposite of above (everything but 'Droid')
df[df['species'] != 'Droid']

# alternative
#df[~(df['species'] == 'Droid')]
```

### 3.2 Select row/columns based on position

```python
# df.iloc(rows, columns)
# : == all
```

```python
# filter first 3 rows, all columns
df.iloc[0:3, :]
```

```python
# filter rows 15 and 17, all columns
df.iloc[[15, 17], :]
```

```python
# filter all rows, first 3 columns
df.iloc[:, 0:3]
```

### 3.3 Select columns based on name

```python
# df.loc(rows, columns)
```

```python
# filter name and eye_color columns
df.loc[:, ['name', 'eye_color']]

# alternative
#df[['name', 'eye_color']]
```

### 3.4 Create new variables and columns

Create 'bmi' column = mass / (height^2)

```python
# check data types
df.dtypes
```

```python
# create 'bmi' column
df['bmi'] = df['mass'] / (df['height'] / 100) ** 2

# round up
df['bmi'] = round(df['bmi'], 2)
df.head()
```

```python
# create. 'bmi_health' column with value 'overweight' if bmi >= 25
df['bmi_health'] = df['bmi'].apply(lambda value: 'overweight' if value >= 25 else 'healthy')
df.head()
```

```python
# alternative all-in-one code
df['bmi_health_aio'] = (
    (round((df['mass'] / (df['height'] / 100) ** 2), 2))
    .apply(lambda value: 'overweight' if value >= 25 else 'healthy')
)

# check if the new columns gives same result
df[['bmi_health', 'bmi_health_aio']]
```

### 3.5 Sort values

```python
# use sort_values by mass in descending order
df.sort_values(by=['mass'], ascending=False).head()
```

### 3.6 Summarize by specific grouping

```python
# mean_mass and count of species
result = df.groupby(by=['species']).agg({
    'mass': 'mean',
    'species': 'count'
}).rename(columns={
    'mass': 'mean_mass',
    'species': 'count'
})

result
```

```python
# show top 3 species
result.sort_values(by=['count'], ascending=False).head(3)
```

### 3.7 Long chain

```python
result.head()
```

```python
# display tallest (> 175) and heaviest (> 100) characters
# group result by homeworld and species
# show avg. height and mass
chain = df[(df['height'] > 175) & (df['mass'] > 100)].groupby(
    by=['homeworld', 'species']).agg({
    'height': 'mean',
    'mass': 'mean'}).rename(
    columns={'height': 'mean_height',
             'mass': 'mean_mass'})

chain
```

### 4.3 With and without outlier

```python
# import matplotlib
import matplotlib

# set figsize
matplotlib.rcParams['figure.figsize'] = (4.8, 3)
```

```python
# without outlier

# plt.xlim([min, max]) to remove outlier
import matplotlib.pyplot as plt

df.plot.scatter(x='mass', y='height')
plt.xlim([0, 200]);
```

```python
# with outlier

# select outlier mass > 1000
outlier =  df[df['mass'] > 1000]

# create variables for outlier mass, height and name
# only 1 row -> row index for values is [0]
outlier_mass = outlier['mass'].values[0]
outlier_height = outlier['height'].values[0]
outlier_name = outlier['name'].values[0]

# create scatter plot with text next to outlier
fig, ax = plt.subplots()
df.plot.scatter(x='mass', y='height', ax=ax)
# add text for value x outlier_mass, y outlier_height, text outlier_name)
ax.text(outlier_mass, outlier_height, outlier_name);
```
